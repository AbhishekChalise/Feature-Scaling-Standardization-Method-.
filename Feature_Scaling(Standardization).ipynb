```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age  = np.random.normal(loc =0 , scale = 10 , size = 1000)
```

```python
salaries = np.random.normal(loc = 0,scale = 10,size = 1000)
salaries
```

```python
data_df = pd.DataFrame({'Age':age,'Salaries':salaries})
data_df
```

```python
data_df
```

```python
sns.histplot(data_df['Age'],kde = True)
```

```python
# Here the Random Distribution is the normal distribution 
# Scaling the values to the desigred range
# First STANDARDIZATION METHOD
```

```python
import sklearn.preprocessing
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standarized_data = scaler.fit_transform(data_df)
```
